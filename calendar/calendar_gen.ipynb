```python
import calendar
import csv

# 设置要生成日历的年份
year = 2023

# 定义要生成日历的月份范围
months = [('July', 7), ('August', 8), ('September', 9), ('October', 10), ('November', 11), ('December', 12)]

# 以写入模式打开 CSV 文件
with open('calendar.csv', 'w', newline='') as file:
    writer = csv.writer(file, delimiter=',')

    # 逐月生成日历并写入 CSV 文件
    for month in months:
        # 获取当前月份的日历数据
        cal = calendar.monthcalendar(year, month[1])

        # 写入月份名
        writer.writerow([month[0]])

        # 写入日历表头
        writer.writerow(['Mo', 'Tu', 'We', 'Th', 'Fr', 'Sa', 'Su'])

        # 写入日历数据
        for row in cal:
            # 格式化每一行中的整数并将它们连接为字符串
            formatted_row = ','.join('{:>2}'.format(day) if day > 0 else '' for day in row)

            # 将格式化的行写入 CSV 文件
            writer.writerow([formatted_row])

        # 写入空行
        writer.writerow([''])
```

```python
import calendar
import xlsxwriter

# 国内假日日期
holidays = [(2023,4,4), (2023,4,5), (2023,4,6), (2023,5,1), (2023,5,28), (2023,5,29), (2023,5,30), (2023,10,1), (2023,10,2), (2023,10,3), (2023,10,4), (2023,10,5), (2023,10,6), (2023,10,7)]

# 创建excel文件
workbook = xlsxwriter.Workbook('calendar.xlsx')
worksheet = workbook.add_worksheet()

# 设置列宽和行高
worksheet.set_column(0, 32, 5)
worksheet.set_default_row(30)

# 设置单元格格式
cell_format = workbook.add_format({'bold': True, 'font_size': 16, 'align': 'center', 'valign': 'vcenter', 'bg_color': '#D9D9D9'})
holiday_format = workbook.add_format({'bg_color': '#FFC7CE'})

# 设置月份、星期和日期的单元格格式
month_format = workbook.add_format({'bold': True, 'font_size': 14, 'align': 'center', 'valign': 'vcenter'})
weekday_format = workbook.add_format({'bold': True, 'font_size': 12, 'align': 'center', 'valign': 'vcenter'})
date_format = workbook.add_format({'bold': False, 'font_size': 12, 'align': 'center', 'valign': 'vcenter'})

# 生成calendar
months = ['April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

row = 0
for month in months:
    # 获取日历月份信息
    cal = calendar.monthcalendar(2023, months.index(month) + 4)

    # 写入月份标题
    worksheet.write(row, 0, month, cell_format)
    worksheet.merge_range(row, 1, row, 7, '', cell_format)
    row += 1

    # 写入星期标题
    weekdays = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
    for weekday in weekdays:
        worksheet.write(row, weekdays.index(weekday) + 1, weekday, weekday_format)
    row += 1

    # 写入日期
    for week in cal:
        for date in week:
            if date == 0:
                worksheet.write(row, week.index(date) + 1, '', date_format)
            else:
                date_str = f"{2023}-{months.index(month)+4}-{date}"
                if (2023, months.index(month)+4, date) in holidays:
                    worksheet.write(row, week.index(date) + 1, date, holiday_format)
                else:
                    worksheet.write(row, week.index(date) + 1, date, date_format)
        row += 1

    # 写入月份之间的空行
    worksheet.merge_range(row, 0, row + 1, 7, '', cell_format)
    row += 2

# 关闭并保存excel文件
workbook.close()
print('Calendar generated and saved as calendar.xlsx')
```

```python
calendar_file = 'calendar1.xlsx'

import calendar
import xlsxwriter
import datetime

# 国内假期日期
holidays = [(2023, 1, 1), (2023, 2, 9), (2023, 2, 10), (2023, 2, 11), (2023, 4, 2), (2023, 4, 3), (2023, 4, 4), (2023, 5, 1), (2023, 6, 2), (2023, 6, 3), (2023, 6, 4), (2023, 9, 30), (2023, 10, 1), (2023, 10, 2), (2023, 10, 3), (2023, 10, 4), (2023, 10, 5), (2023, 10, 6), (2023, 10, 7)]

# 官方公布的调休
makeups = [(2023, 4, 1), (2023, 4, 5), (2023, 9, 29)]

# 创建excel文件
workbook = xlsxwriter.Workbook(calendar_file)
worksheet = workbook.add_worksheet()

# 设置列宽和行高
worksheet.set_column(0, 32, 5)
worksheet.set_default_row(30)

# 设置单元格格式
cell_format = workbook.add_format({'bold': True, 'font_size': 16, 'align': 'center', 'valign': 'vcenter', 'bg_color': '#D9D9D9'})
holiday_format = workbook.add_format({'bg_color': '#FFC7CE'})

# 设置月份、星期和日期的单元格格式
month_format = workbook.add_format({'bold': True, 'font_size': 14, 'align': 'center', 'valign': 'vcenter'})
weekday_format = workbook.add_format({'bold': True, 'font_size': 12, 'align': 'center', 'valign': 'vcenter'})
date_format = workbook.add_format({'bold': False, 'font_size': 12, 'align': 'center', 'valign': 'vcenter'})

# 生成calendar
months = ['April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

row = 0
makeup_len = len(makeups)
for month in months:
    # 获取日历月份信息
    cal = calendar.monthcalendar(2023, months.index(month) + 4)

    # 写入月份标题
    worksheet.write(row, 0, month, cell_format)
    worksheet.merge_range(row, 1, row, 7, '', cell_format)
    row += 1

    # 写入星期标题
    weekdays = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
    for weekday in weekdays:
        worksheet.write(row, weekdays.index(weekday) + 1, weekday, weekday_format)
    row += 1

    # 写入日期
    for week in cal:
        for date in week:
            if date == 0:
                worksheet.write(row, week.index(date) + 1, '', date_format)
            else:
                date_time = datetime.date(2023, months.index(month) + 4, date)
                date_str = date_time.strftime('%Y-%m-%d')
                if (2023, months.index(month) + 4, date) in holidays:
                    worksheet.write(row, week.index(date) + 1, date, holiday_format)
                elif (2023, months.index(month) + 4, date) in makeups:
                    worksheet.write(row, week.index(date) + 1, date, holiday_format)
                elif week.index(date) == 5 or week.index(date) == 6:  # 周六日加粗
                    worksheet.write(row, week.index(date) + 1, date, date_format)
                else:
                    worksheet.write(row, week.index(date) + 1, date, date_format)
        row += 1

    # 写入月份之间的空行
    if (months.index(month) + 4) != 12:
        worksheet.merge_range(row, 0, row + 1, 7, '', cell_format)
        row += 2

    # 处理调休
    for makeup in makeups:
        if makeup[1] == (months.index(month) + 4):
            if (makeup[2] >= 1 and makeup[2] <= 7):
                makeup_date = datetime.date(makeup[0], makeup[1], makeup[2])
                weekday = makeup_date.weekday()
                week_num = makeup_date.day // 7 + 1
                worksheet.write(row - 9 + week_num, weekday + 1, makeup[2], holiday_format)
            elif (makeup[2] >= 25 and makeup[2] <= 30):
                makeup_date = datetime.date(makeup[0], makeup[1], makeup[2])
                weekday = makeup_date.weekday()
                worksheet.write(row, weekday + 1, makeup[2], holiday_format)

# 关闭并保存excel文件
workbook.close()
print(f'Calendar generated and saved as {calendar_file}')
```
