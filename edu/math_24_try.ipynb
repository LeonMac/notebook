```python
import operator
import itertools
import sys

sys.set_int_max_str_digits(5000)  # 例如，设置为 5000 位

# 定义支持的操作符和相应的函数
OPS = {
    '+': operator.add,
    '-': operator.sub,
    '*': operator.mul,
    '/': operator.truediv,
    # '**': operator.pow
}

# 检查表达式是否为目标值
def check_expression(inputs, target):
    n_try = 0
    for ops in itertools.product(OPS.keys(), repeat=len(inputs)-1):
        for perm in itertools.permutations(inputs):
            expr = build_expression(perm, ops)
            
            try:
                print(f"try [{n_try}] {expr} = ", end = '')
                result = eval(expr)
                print(f"{result}")
                if result == target:
                    return expr
            except (ZeroDivisionError, SyntaxError):
                pass  # 忽略除以零的错误和语法错误

            n_try += 1
    return None

# 根据运算符构建表达式字符串
def build_expression(inputs, ops):
    # 确保运算符数量和输入数量匹配
    if len(ops) != len(inputs) - 1:
        return ""
    # 构建表达式
    expr_parts = [str(inputs[0])]
    for op, num in zip(ops, inputs[1:]):
        expr_parts.append(f"{op}({num})")
    return " ".join(expr_parts)


# 回溯函数来构建表达式
# 回溯函数来构建表达式
def backtrack(index, expr, values, target):
    # 基础条件：如果达到索引的末尾且表达式结果等于目标，则返回表达式
    if index == len(values):
        return str(eval(expr[-1])) if expr else None
    
    # 初始化当前表达式结果
    
    current_result = expr[-1] if expr else 0

    # 尝试当前索引的数字直接作为结果
    new_expr = expr + [str(values[index])]
    if eval(new_expr[-1]) == target:
        return new_expr[-1]

    # 遍历所有操作符
    for op in OPS:
        for prev_index in range(index):
            # 构建新的表达式片段
            new_expr = expr[:expr.index(str(values[prev_index]))+1] + [op] + new_expr
            # 计算新表达式的结果
            try:
                new_result = str(OPS[op](current_result, values[index]))
                # 如果新结果是目标，则返回这个表达式
                if eval(new_result) == target:
                    return new_expr + [new_result]
                # 否则，将其添加到表达式列表中并继续回溯
                current_result = eval(new_expr[-1])
            except ZeroDivisionError:
                continue  # 忽略除以零的错误

    return None


# 输入的整数列表和目标输出整数
inputs = [20, 20, 9, 9, 7]
target = 35

# 查找表达式
# expression = check_expression(inputs, target)

# 调用回溯函数
expression = backtrack(0, [], inputs, target)

# 输出结果
if expression:
    print(f"找到了一个表达式: {expression} = {target}")
else:
    print("没有找到匹配的表达式。")
```
