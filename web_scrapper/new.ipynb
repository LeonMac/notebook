```python
import requests
from bs4 import BeautifulSoup
```

```python
HEADER = {'User-Agent': 'Mozilla/5.0 (iPhone; CPU iPhone OS 11_0 like Mac OS X) AppleWebKit'}

TOP_PAGE = 'https://h.fkxs.net/book/236625/115540308.html'
```

```python
r = requests.get(TOP_PAGE, headers=HEADER)
# 返回请求状态码，200即为请求成功
print(f"request status = {r.status_code}")
```

```python
# 返回页面代码
print(r.text)
```

```python
# 对于特定类型请求，如Ajax请求返回的json数据
# print(r.json())
```

```python
# 保持会话
# 新建一个session对象
sess = requests.session()
# # 先完成登录
# sess.post('maybe a login url', data=data, headers=HEAD)
# 然后再在这个会话下去访问其他的网址
r = sess.get(TOP_PAGE, headers=HEADER)
```

```python
html = r.text
```

```python
# 选用lxml解析器来解析html
soup = BeautifulSoup(html, 'lxml')
```

```python
# 获取标题
print(soup.title)
```

```python
# 获取文本
print(soup.title.text)
```

```python
# 通过标签定位
div_element = soup.find('div', {'id': 'nr'})

# 提取出div内的所有内容
div_contents = div_element.text

# 输出div内容
print(div_contents)
```

```python
# 通过属性定位
print(soup.find_all(attrs={'id': 'link1'}))
```

```python
# 标签 + 属性定位
print(soup.find_all('a', id='link1'))
```
